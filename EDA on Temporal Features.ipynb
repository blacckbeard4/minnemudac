```python
from scipy.stats import ks_2samp, kruskal
# Sample dataset
#df_novice_length=df_novice_length[df_novice_length['Stage']=='Closed']
df_novice_length=df_novice[['Match Activation Date','Match Length','Stage']]
df = df_novice_length

# Compute end date
df['end_date'] = df['Match Activation Date'] + pd.to_timedelta(df['Match Length'] * 30, unit='D')

# Define time bins (e.g., by year)
df['year'] = df['Match Activation Date'].dt.year

# Plot distributions over years
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['year'], y=df['Match Length'])
plt.xticks(rotation=45)
plt.title("Response Variable Distribution Over Years")
plt.show()

# Compare distributions statistically (e.g., last 5 years vs. first 5 years)
early_period = df[df['year'] < df['year'].median()]['Match Length']
late_period = df[df['year'] >= df['year'].median()]['Match Length']

# Kolmogorov-Smirnov Test (checks if two distributions differ)
ks_stat, ks_p = ks_2samp(early_period, late_period)
print(f"KS Test p-value: {ks_p:.4f} (p < 0.05 suggests significant difference)")

# Kruskal-Wallis Test (non-parametric ANOVA for multiple years)
kruskal_stat, kruskal_p = kruskal(*[df[df['year'] == y]['Match Length'] for y in df['year'].unique()])
print(f"Kruskal-Wallis Test p-value: {kruskal_p:.4f} (p < 0.05 suggests significant difference)")
```

```python
df_novice_length_churn = df_novice[['Match Activation Date', 'Match Length', 'Stage']]
df_novice_length_churn = df_novice_length_churn[df_novice_length_churn['Stage'] == 'Closed']
df_novice_length_churn = df_novice_length_churn[df_novice_length_churn['Match Length'] <= 12]

# Calculate overall churn percentage
churn_percentage = df_novice_length_churn.shape[0] / df_novice.shape[0]
print('Percentage of 1-year churn:', churn_percentage)


# Convert to datetime
df_novice_length_churn['year'] = df_novice_length_churn['Match Activation Date'].dt.year

# Calculate total number of matches per year
total_matches_per_year = df_novice.groupby(df_novice['Match Activation Date'].dt.year).size()

# Calculate churned matches per year
yearly_counts = df_novice_length_churn['year'].value_counts().sort_index()

# Calculate the percentage of churned matches per year
yearly_percentage = (yearly_counts / total_matches_per_year) * 100  # Normalize by total matches per year

# Plot bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x=yearly_percentage.index, y=yearly_percentage.values, palette="Blues_r")

# Formatting
plt.title("Year-wise Percentage of One-Year Churned Matches")
plt.xlabel("Year")
plt.ylabel("Churn Percentage (%)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
